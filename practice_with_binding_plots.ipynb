## Hyperbolic
For the reaction $P+L\rightleftharpoons PL$, given an equilibrium dissociation constant, $$K_D=\frac{[P][L]}{[PL]}$$

where $[P]$ and $[L]$ refer to the concenrations *free* the unbound components, and $[PL]$ is the concentration of their complex,

the fraction bound has a hyperbolic shape with respect to $[L]$:
$$Y = \frac{[PL]}{[P]+[PL]} = \frac{1}{\frac{[P]}{[PL]} + 1}= \frac{1}{\frac{K_D}{[L]} + 1}=\frac{[L]}{K_D+[L]}$$



## Quadratic
For the reaction $P+L\rightleftharpoons PL$, given total concentrations of P and L, and a dissociation constant $K_D$, in the general case, we use mass balance to account for both the free and bound species of the components.

Fraction bound: $$Y=\frac{[PL]}{[P_{tot}]}$$

Bound concentration is determined by applying mass balance to the definition of the equilibrium dissociation constant, $K_D$:

$$K_D = \frac{[P][L]}{[PL]}=\frac{([P_{tot}]-[PL])([L_{tot}]-[PL])}{[PL]}$$

This can be rearragned to:

$$0 = [PL]^2 -([L_{tot}]+[P_{tot}]+K_D)[PL]+[P_{tot}][L_{tot}]$$

which has the general form of a quadratic equation $ax^2 + bx + c = 0$ where $x=[PL]$, and can be solved using:

$$x=\frac{-b-\sqrt{b^2-4ac}}{2a}$$

The solution of this equation tells us the concentration of [PL] as a funtion of the *total* concentration of components, and the dissociation constant:

$$[PL] = \frac{[P_{tot}]+[L_{tot}]+K_D-\sqrt{([P_{tot}]+[L_{tot}]+K_D)^2-4[P_{tot}][L_{tot}]}}{2}$$

Divide that by $[P_{tot}]$ and we have $Y$.
